### Q2. Write a python program to print Python Exception Hierarchy.

```python
def print_exception_hierarchy(exception_class, level=0):
    print("    " * level + exception_class.__name__)
    for subclass in exception_class.__subclasses__():
        print_exception_hierarchy(subclass, level + 1)

print_exception_hierarchy(Exception)
```

```python
# Print the hirrarcy of ArithemeticError Exception
print_exception_hierarchy(ArithmeticError)
```

### Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

> The `LookupError` class is a base class for exceptions that are raised when an indexing or key lookup operation fails in a dictionary or list.

```python
# KeyError is raised when a dictionary does not have a key that is being looked up. 

d = {"a": 1, "b": 2}
try:
    print(d["c"])
except KeyError as e:
    print("KeyError:", e)
```

```python
# IndexError is raised when a list index is out of range.

l = [1, 2, 3]
try:
    print(l[3])
except IndexError as e:
    print("IndexError:", e)
```

### Q5. Explain ImportError. What is ModuleNotFoundError?

> `ImportError` occurs when you try to import modules or pakages which are not _present_ or _installed_ in the `sys.path` or `PYTHONPATH`.

> `ModuleNotFoundError` occurs when a module cannot be found in the `sys.path` or `PYTHONPATH`.

### Q6. List down some best practices for exception handling in python.

#### Some best practices for exception handling:

1. Always catch specific exceptions instead of general `Exception` class.
2. Provide meaningful error messages to the user with `print` statement or by `logging` method.
3. Handel exceptions in right order means order your `except` blocks from the most specific to the most general.This helps in avoiding unexpected behavior and ensures that the right exception is caught.
4. Use `logging` instead of `print` because it helps in debugging process of code.
